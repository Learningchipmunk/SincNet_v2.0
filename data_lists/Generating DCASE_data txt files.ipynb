# This notebook is purely for re-writing the fold1 txt files


```python
import pandas as pd
import numpy as np
```

### Reading the files and storing them into data frames:

```python
trainFilesPath = "fold1_train.txt"
validFilesPath = "fold1_valid.txt"

trainDataFrame = pd.read_csv(trainFilesPath, delimiter="\t", names = ["File Name", "Label", "Manually Verified"])
validDataFrame = pd.read_csv(validFilesPath, delimiter="\t", names = ["File Name", "Label", "Manually Verified"])

# Showing 47 rows:
print(trainDataFrame.iloc[:47])
print(validDataFrame.iloc[:47])
```


### An example of path extraction here: 

```python
print("Initial String:")
print(trainDataFrame['File Name'][1])
print("\n")
print("After Partitionning:")
print(trainDataFrame['File Name'][1].partition('/'))
print("\n")
print("Final form:")
print(trainDataFrame['File Name'][1].partition('/') [-1])
```

```python
outputTrainFileName = "DCASE_train.txt"
outputValidFileName = "DCASE_valid.txt"


for path, dataFrame in ([[outputTrainFileName, trainDataFrame], [outputValidFileName, validDataFrame]]):
    print(dataFrame)
    file = open(path,"w+") 
    
    for index, row in dataFrame.iterrows():
        file.write(row['File Name'].partition('/') [-1] + "\n") 
    
    file.close()
```
